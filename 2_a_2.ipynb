```python
import pandas as pd
import numpy as np
```

```python
position_lable = pd.read_csv('./data/positoin_label.csv').drop('Unnamed: 0', axis=1)
position_lable
```

```python
position_year_average_temp = pd.read_csv('./data/position_year_average_temp.csv').drop('Unnamed: 0', axis=1)
position_year_average_temp['Lable'] = None
position_year_average_temp
```

```python
position_year_average_temp_label = pd.DataFrame(columns=position_year_average_temp.columns)
for i in range(6):
    cities = position_lable.loc[position_lable['Label']==i]['City'].to_list()
    temp = position_year_average_temp[position_year_average_temp['City'].isin(cities)]
    temp['Lable'] = i
    position_year_average_temp_label = pd.concat([position_year_average_temp_label, temp])
```

```python
position_year_average_temp_label.to_csv('./data/position_year_average_temp_label.csv', index=None)
position_year_average_temp_label
```

```python
city0_df = position_year_average_temp_label.loc[position_year_average_temp_label['Lable']==0].reset_index(drop=True)
city1_df = position_year_average_temp_label.loc[position_year_average_temp_label['Lable']==1].reset_index(drop=True)
city2_df = position_year_average_temp_label.loc[position_year_average_temp_label['Lable']==2].reset_index(drop=True)
city3_df = position_year_average_temp_label.loc[position_year_average_temp_label['Lable']==3].reset_index(drop=True)
city4_df = position_year_average_temp_label.loc[position_year_average_temp_label['Lable']==4].reset_index(drop=True)
city5_df = position_year_average_temp_label.loc[position_year_average_temp_label['Lable']==5].reset_index(drop=True)
```

```python
city0_df.to_csv('./data/city0_df.csv', index=None)
```
