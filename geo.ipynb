```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv('./data/position_year_average_temp_label.csv').drop_duplicates('City')
df
```

```python
city0 = df.loc[df['Lable']==0]
city0
```

```python
city1 = df.loc[df['Lable']==1]
city1
```

```python
city2 = df.loc[df['Lable']==2]
city2
```

```python
city3 = df.loc[df['Lable']==3]
city3
```

```python
city4 = df.loc[df['Lable']==4]
city4
```

```python
city5 = df.loc[df['Lable']==5]
city5
```

```python
gdf0 = geopandas.GeoDataFrame(
    city0, geometry=geopandas.points_from_xy(-city0.Longitude, city0.Latitude))
gdf1 = geopandas.GeoDataFrame(
    city1, geometry=geopandas.points_from_xy(-city1.Longitude, city1.Latitude))
gdf2 = geopandas.GeoDataFrame(
    city2, geometry=geopandas.points_from_xy(-city2.Longitude, city2.Latitude))
gdf3 = geopandas.GeoDataFrame(
    city3, geometry=geopandas.points_from_xy(-city3.Longitude, city3.Latitude))
gdf4 = geopandas.GeoDataFrame(
    city4, geometry=geopandas.points_from_xy(-city4.Longitude, city4.Latitude))
gdf5 = geopandas.GeoDataFrame(
    city5, geometry=geopandas.points_from_xy(-city5.Longitude, city5.Latitude))
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))

# We restrict to South America.
ax = world.plot(
     color='white',
     edgecolor='black',figsize=(30,30))

# We can now plot our ``GeoDataFrame``.
gdf0.plot(ax=ax, color='g', label='Cities 0')
gdf1.plot(ax=ax, color='k', label='Cities 1')
gdf2.plot(ax=ax, color='b', label='Cities 2')
gdf3.plot(ax=ax, color='c', label='Cities 3')
gdf4.plot(ax=ax, color='m', label='Cities 4')
gdf5.plot(ax=ax, color="#FF9900", label='Cities 5')
plt.legend(prop={'size': 20})
plt.title("Geographical distribution of 6 types of cities", fontsize=30)
plt.show()
```
