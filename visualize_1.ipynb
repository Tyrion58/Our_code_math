```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
```

```python
global_data = pd.read_csv('./data/global_data.csv')
global_data.drop('Unnamed: 0', axis=1, inplace=True)
global_data
```

```python
plt.plot(global_data['dt'], global_data['mean_temp'], color='r')
plt.ylabel('global temperature')
plt.xlabel('year')
plt.title('Global average temperature variation')
plt.show()
```

```python
plt.hist(global_data['mean_temp'], bins=20, edgecolor='b')
plt.xlabel("temperature")
plt.title("Global average temperature frequency")
plt.show()
```
